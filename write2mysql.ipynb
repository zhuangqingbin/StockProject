```python
dates = pd.date_range(start='20100101', end='20191231', freq='D')
date_list = dates.strftime('%Y%m%d').tolist()  # 转为字符串列表
print(date_list[:5])  # 查看前5个日期
print(f"共计 {len(date_list)} 天")

engine = create_engine("mysql+pymysql://root:zqb123456@localhost/stock_data?charset=utf8")
for date in date_list:
    time.sleep(0.5)
    stock_daily.fetch_data("当天数据", trade_date = date)
    df = stock_daily.data
    if len(df) > 0:
        print(date)
        df['trade_date'] = pd.to_datetime(df['trade_date'])
        df.to_sql('daily_kline', engine, if_exists='append', index=False, method='multi')
```
